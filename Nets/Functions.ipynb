```python
import torch
import numpy as np
import matplotlib.pyplot as plt
torch.cuda.is_available()
```

```python
# 激活函数
## Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100) # 从-10到10生成100个点
y = sigmoid(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="sigmoid", color="blue")
plt.title("Sigmoid Function")
plt.xlabel("Input")
plt.ylabel("Output")
# 显示网格线
plt.grid(True, which='both', axis='both', linestyle='--')
# 显示图例，即在plot中设置的label
plt.legend()
plt.savefig("assets/sigmoid.png")
```

```python
## tanh 双曲正切函数
def tanh(x):
    return 2 / (1 + np.exp(-2 * x)) - 1

x = np.linspace(-10, 10, 100)
y = tanh(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="tanh", color="blue")
plt.title("Tanh Funtion")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True, which='both', axis='both', linestyle='--')
plt.legend()
plt.savefig("assets/tanh.png")
```

```python
## ReLU
def ReLU(x):
    return np.maximum(0, x)
    
x = np.linspace(-10, 10, 100)
y = ReLU(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="ReLU", color="blue")
plt.title("ReLU Funtion")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True, which='both', axis='both', linestyle='--')
plt.legend()
plt.savefig("assets/ReLU.png")
```

```python
## Leaky ReLU
def LeakyReLU(x, alpha=0.1):
    return np.where(x>=0, x, alpha * x)
    
x = np.linspace(-10, 10, 100)
y = LeakyReLU(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Leaky ReLU", color="blue")
plt.title("Leaky ReLU Funtion")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True, which='both', axis='both', linestyle='--')
plt.legend()
plt.savefig("assets/Leaky ReLU.png")
```

```python
## ELU
def ELU(x, alpha=0.1):
    return np.where(x>0, x, alpha * (np.exp(x) - 1))
    
x = np.linspace(-10, 10, 100)
y = ELU(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="ELU", color="blue")
plt.title("ELU Funtion")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True, which='both', axis='both', linestyle='--')
plt.legend()
plt.savefig("assets/ELU.png")
```

```python
# 对比
x = np.linspace(-10, 10, 100)
ReLUY = ReLU(x)
LeakyY = LeakyReLU(x, 0.1)
ELUY = ELU(x, 0.1)

plt.figure(figsize=(8, 6))
plt.plot(x, ReLUY, label="ReLU", color="blue")
plt.plot(x, LeakyY, label="Leaky ReLU", color="green")
plt.plot(x, ELUY, label="ELU", color="red")
plt.title("Comparation")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True, which='both', axis='both', linestyle='--')
plt.legend()
plt.savefig("assets/relu compare.png")
```
